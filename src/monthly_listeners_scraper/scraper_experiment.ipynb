```python
driver = webdriver.Chrome()
driver.get('https://open.spotify.com/album/3c0NHNo2Gn0X7uARad3hGv')
```

#get links for all songs in the album in a list
#extract streams after going into each link

```python
# track_elements = driver.find_elements(By.CSS_SELECTOR, 'a[data-testid="internal-track-link"]')

#for playlist
track_elements = driver.find_elements(By.CSS_SELECTOR, 'div[role="grid"][data-testid="playlist-tracklist"] a[data-testid="internal-track-link"]')
```

```python
track_elements
```

```python
track_links = [track.get_attribute('href') for track in track_elements]
```

```python
track_links
```

```python
track_streams = {}
name = []
plays = []

for link in track_links:
    driver.get(link)
    time.sleep(5)

    name.append(driver.find_element(By.CSS_SELECTOR, 'h1[data-encore-id="text"]').text)
    plays.append(driver.find_element(By.CSS_SELECTOR, 'span[data-testid="playcount"]').text)
    
```

```python
driver.quit()
```

```python
track_streams['Name'] = name
track_streams['Plays'] = plays
track_streams
```

```python
track_streams['Name'] = name
track_streams['Plays'] = plays
```

```python
with open('drake_album_data.csv','w',newline='') as f:
    writer = csv.writer(f)
    writer.writerow(track_streams.keys())
    for i in range(len(name)):
        writer.writerow([val[i] for val in track_streams.values()])
```
